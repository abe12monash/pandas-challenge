```python
# Scores by School Spending
# Define a function to calculate the percentage of students passing math and reading
def passing_percentage(passing_students, total_students):
    return passing_students / total_students * 100

# Calculate the average math and reading scores
average_math_score = school_data_complete['maths_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

# Calculate the number of students passing math and reading (assuming passing is 70 or greater)
passing_math = school_data_complete[school_data_complete['maths_score'] >= 50]
passing_reading = school_data_complete[school_data_complete['reading_score'] >= 50]

# Calculate the passing rates
passing_math_percentage = passing_percentage(passing_math['Student ID'].nunique(), school_data_complete['Student ID'].nunique())
passing_reading_percentage = passing_percentage(passing_reading['Student ID'].nunique(), school_data_complete['Student ID'].nunique())

# Calculate the overall passing rate (average of the math and reading passing rates)
overall_passing_rate = (passing_math_percentage + passing_reading_percentage) / 2
# Define bins for spending
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

# Calculate per student budget
school_data_complete['Per Student Budget'] = school_data_complete['budget'] / school_data_complete['size']

# Segment and sort data into bins based on the per student budget
school_data_complete['Spending Ranges (Per Student)'] = pd.cut(school_data_complete['Per Student Budget'], bins, labels=group_names)

# Perform groupby on 'Spending Ranges (Per Student)' and calculate averages and passing percentages
spending_stats = school_data_complete.groupby('Spending Ranges (Per Student)')

# Calculate averages for the desired columns.
average_math_score = spending_stats['maths_score'].mean()
average_reading_score = spending_stats['reading_score'].mean()

# Calculate the percentage of students passing math and reading
passing_math_count = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('Spending Ranges (Per Student)').count()['maths_score']
passing_reading_count = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('Spending Ranges (Per Student)').count()['reading_score']
total_students = spending_stats['Student ID'].count()

passing_math_percentage = passing_math_count / total_students * 100
passing_reading_percentage = passing_reading_count / total_students * 100

# Calculate overall passing rate
# use and
overall_passing_rate = school_data_complete[(school_data_complete['reading_score'] >= 50) & (school_data_complete['maths_score'] >= 50)].groupby('Spending Ranges (Per Student)').count()['Student ID'] / total_students * 100



# Create a summary DataFrame
spending_summary_df = pd.DataFrame({
    "Average Maths Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Maths": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Passing": overall_passing_rate
})

# Formatting the DataFrame
spending_summary_df['Average Maths Score'] = spending_summary_df['Average Maths Score'].map("{:.2f}".format)
spending_summary_df['Average Reading Score'] = spending_summary_df['Average Reading Score'].map("{:.2f}".format)
spending_summary_df['% Passing Maths'] = spending_summary_df['% Passing Maths'].map("{:.2f}".format)
spending_summary_df['% Passing Reading'] = spending_summary_df['% Passing Reading'].map("{:.2f}".format)
spending_summary_df['% Overall Passing'] = spending_summary_df['% Overall Passing'].map("{:.2f}".format)

spending_summary_df
```

```python
# Define bins for school size
size_bins = [0, 1000, 2000, 5000]
size_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Segment and sort data into bins based on school size
school_data_complete['School Size'] = pd.cut(school_data_complete['size'], size_bins, labels=size_group_names)

# Perform groupby on 'School Size' and calculate averages and passing percentages
size_stats = school_data_complete.groupby('School Size')

# Calculate averages for the desired columns.
average_math_score_size = size_stats['maths_score'].mean()
average_reading_score_size = size_stats['reading_score'].mean()

# Calculate the percentage of students passing math and reading
passing_math_count_size = school_data_complete[school_data_complete['maths_score'] >= 50].groupby('School Size').count()['maths_score']
passing_reading_count_size = school_data_complete[school_data_complete['reading_score'] >= 50].groupby('School Size').count()['reading_score']
total_students_size = size_stats['Student ID'].count()

passing_math_percentage_size = passing_math_count_size / total_students_size * 100
passing_reading_percentage_size = passing_reading_count_size / total_students_size * 100

# Calculate overall passing rate
overall_passing_rate_size = school_data_complete[(school_data_complete['reading_score'] >= 50) & 
                                                 (school_data_complete['maths_score'] >= 50)].groupby('School Size').count()['Student ID'] / total_students_size * 100

# Create a summary DataFrame
size_summary_df = pd.DataFrame({
    "Average Maths Score": average_math_score_size,
    "Average Reading Score": average_reading_score_size,
    "% Passing Maths": passing_math_percentage_size,
    "% Passing Reading": passing_reading_percentage_size,
    "% Overall Passing": overall_passing_rate_size
})

# Formatting the DataFrame
size_summary_df['Average Maths Score'] = size_summary_df['Average Maths Score'].map("{:.6f}".format)
size_summary_df['Average Reading Score'] = size_summary_df['Average Reading Score'].map("{:.6f}".format)
size_summary_df['% Passing Maths'] = size_summary_df['% Passing Maths'].map("{:.6f}".format)
size_summary_df['% Passing Reading'] = size_summary_df['% Passing Reading'].map("{:.6f}".format)
size_summary_df['% Overall Passing'] = size_summary_df['% Overall Passing'].map("{:.6f}".format)

size_summary_df
```
